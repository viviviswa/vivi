### MULTIVARIATE ANALYSIS

### Line plot

```python
plt.figure(figsize=(6,3))
sns.lineplot('Age','Fare',color='green',data=data)
plt.title("Age vs Fare")
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.lineplot('Age','Class',color='blue',data=data)
plt.title("Age vs Class")
plt.show()
```

### PIE CHART

```python
data.columns
```

```python
Pclass_survived=data.groupby(['Pclass'])['Survived'].sum()
```

```python
Pclass_survived
```

```python
sns.set_style('ticks')
Pclass_survived.plot.pie()
plt.legend()
plt.show()
```

```python
Pclass_sex_survived=data.groupby(['Pclass','Sex'])['Survived'].sum()
Pclass_sex_survived
```

```python
sns.set_style('ticks')
Pclass_sex_survived.plot.pie()
plt.legend(bbox_to_anchor=(1.5,1),loc='upper left',borderaxespad=10)
plt.show()
```

```python
Pclass_sex_survived.plot.pie(autopct='%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1),loc='upper left',borderaxespad=20)
plt.show()
```
