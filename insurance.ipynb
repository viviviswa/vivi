Discriptive Method

In which region there is maximum amount of smokers present?

```python
data_smoker=data.query("smoker=='yes'")
```

```python
data_smoker.region.value_counts()
```

how many males are present below age 18 and smoker also?

```python
sex_age_smoker=data.query("age<19 & sex=='male'& smoker=='yes'")
```

```python
sex_age_smoker.value_counts()
```

How many males present in each northwest region and are smokers?

```python
male_reg_smokers=data.query("sex=='male' & region=='northwest'& smoker=='yes'" )
```

```python
male_reg_smokers.value_counts()
```

How many females below 30 year old present in each southeast region,southwest and are smokers?

```python
se_sw_fesmoker1=data.query("sex=='female' & smoker=='yes'& age<30 & region==['southeast','southwest']")
```

```python
se_sw_fesmoker1
```
